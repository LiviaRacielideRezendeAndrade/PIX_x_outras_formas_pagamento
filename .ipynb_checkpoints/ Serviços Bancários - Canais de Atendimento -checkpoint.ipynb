```python
# Fonte de dados Banco central.Volumetria por canal de acesso e  produto
url ="https://olinda.bcb.gov.br/olinda/servico/MPV_DadosAbertos/versao/v1/odata/TRANSOPADA(trimestre=@trimestre)?@trimestre='20101'&$top=10000&$format=text/csv&$select=trimestre,canalAcesso,produto,acessoATM,qtdTransacoes,valorTransacoes" 
response = requests.get(url)

if response.status_code == 200:
    print ('Sucesso')
else:
    print(f"Erro na requisição: {response.status_code}")
    
font = pd.read_csv(url)
font
```

```python
font['valorTransacoes'] = font['valorTransacoes'].astype(str).str.replace(',', '.')
font['valorTransacoes'] = font['valorTransacoes'].astype(float)
grouped = font.groupby(['trimestre', 'canalAcesso', 'produto'])['valorTransacoes'].sum().reset_index()
grouped
```

```python
a= font.groupby('produto')['valorTransacoes'].sum().sort_values(ascending=True)
a
```

### Evolução das principais transações em relação ao tempo

```python
font['produto'].unique()
```

```python
df=font.copy()
# Garante que o valor seja string para poder extrair partes
df['trimestre'] = df['trimestre'].astype(str)

# Extrai ano e número do trimestre
df['ano'] = df['trimestre'].str[:4].astype(int)
df['n_trimestre'] = df['trimestre'].str[4:].astype(int)

# Cria coluna 'data' com a data correspondente ao início do trimestre
df['data'] = pd.to_datetime(df['ano'].astype(str) + 'Q' + df['n_trimestre'].astype(str))
df
```

### Depósito x Canal

```python
produto_escolhido = 'Depósito'
df_filtro = df[df['produto'] == produto_escolhido]

# Agrupa por data e canal
df_agrupado = df_filtro.groupby(['data', 'canalAcesso'])['valorTransacoes'].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_agrupado, x='data', y='valorTransacoes', hue='canalAcesso', marker='o')
plt.title(f"Evolução do valor das transações por canal no produto: {produto_escolhido}")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Transações realizadas por equipamento próprio do cliente

```python
font['canalAcesso'].unique()
```

```python
canais_g = font.groupby(['trimestre', 'canalAcesso'])['valorTransacoes'].sum().reset_index()
# Formata o valor das transações com separador de milhar
canais_g['valorTransacoes'] = canais_g['valorTransacoes'].apply(lambda x: f"{x:,.0f}")
canais_g
```

```python
canais_g['canal'] = np.where(
    canais_g['canalAcesso'].isin(['Telefone Celular', 'Internet Banking']),
    'celular_internet',
    'rede_bancária'
)
canais_g = canais_g[canais_g['canal'] != 'Iniciador de Transações de Pagamentos']
canais_g= canais_g[['trimestre','canal','valorTransacoes']]
```

```python
canais= canais_g.copy()
canais['valorTransacoes'] = canais['valorTransacoes'].replace({',': ''}, regex=True).astype(float)
canais = canais.groupby(['trimestre', 'canal'])['valorTransacoes'].sum().unstack()

canais.plot(kind='bar', figsize=(20, 6))
plt.title('Evolução dos Canais Digitais ao Longo dos Trimestres')
plt.xlabel('Trimestre')
plt.ylabel('Valor das Transações')
plt.xticks(rotation=45)
plt.legend(title='Canal')
plt.tight_layout()
plt.show()
```

### Crescimento do Canal Celular/Internet
O uso dos canais digitais como celular e Internet_Banking apresentou um crescimento significativo ao longo dos anos, refletindo a crescente adoção das novas tecnologias pelos usuários para realizar transações financeiras.
No início de 2022 ocorre a inversão do uso dos canais digitais.
